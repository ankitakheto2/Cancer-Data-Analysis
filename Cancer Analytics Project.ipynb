```python
#change datatype
import datetime as dt
df["Diagnosis_Date"]=pd.to_datetime(df["Diagnosis_Date"])
df["Treatment_Start_Date"]=pd.to_datetime(df["Treatment_Start_Date"])
df['Death_Date'] = pd.to_datetime(df['Death_Date'])

df.dtypes
```

```python
#create "Age Category "column
print(df["Age"].describe())
#create bins for the verious age category
min_age=df["Age"].min()
twenty_fifth=df["Age"].quantile(0.25)
median=df["Age"].median()
seventy_fifth=df["Age"].quantile(0.75)
max_age=df["Age"].max()
#create a label and bins for the verious age category
labels=["young_adult","middle_age","older_adult","elderlist"]
bins=[min_age,twenty_fifth,median,seventy_fifth,max_age]
#create a category column for age
df["Age_category"]=pd.cut(df["Age"],labels=labels,bins=bins)
```

```python
df[["Age","Age_category"]]
```

```python
#find which patient are death
df[df["Outcome"]=="Deceased"]
```

```python
#fill the null values
df['Comorbidities'] = df['Comorbidities'].fillna('Unknown')
df
```

```python
# for alive
df.loc[df['Outcome'] != 'Deceased', 'Cause_of_Death'] = 'Not Applicable'

```

```python
df
```

```python
#Cheack missing values 
df.isnull().sum()
```

```python
#create column
df['Patient_Status']=np.where((df['Outcome']=='Deceased'),"Dead","Alive")
df
```

```python
# BMI
df['BMI'] = df['Weight'] / ((df['Height']/100)**2)

# Survival Days
df['Survival_Days'] = (df['Death_Date'].fillna(df['Treatment_Start_Date']) - df['Diagnosis_Date']).dt.days

# Year column
df['Diagnosis_Year'] = df['Diagnosis_Date'].dt.year
df
```

```python
#Alive vs Dead
df["Patient_Status"].value_counts().plot(kind='bar')
plt.title("Alive Vs Dead Patients")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
#Cancer Type Distribution
df['Cancer_Type'].value_counts().plot(kind='bar')
plt.title("Cancer Type Distribution")
plt.xlabel("Cancer Type")
plt.ylabel("Count")
plt.show()
```

```python
#Cancer Stage Distribution
df["Cancer_Stage"].value_counts().plot(kind='bar',color='r')
plt.title("Cancer Stage Distribution")
plt.show()
```

```python
#Comorbidities
df["Comorbidities"].value_counts().plot(kind='bar',color='lightgreen')
plt.title("Comorbidities")
plt.show()
```

```python
#Gender Distribution
df["Gender"].value_counts().plot(kind='pie',autopct="%1.1f%%",startangle=45)
plt.title("Gender Distribution")
plt.show()
```

```python
#Ethnicity Distribution
df["Ethnicity"].value_counts().plot(kind='barh')

counts = df["Ethnicity"].value_counts()

for i, v in enumerate(counts):
    plt.text(v, i, str(v), va='center')

plt.title("Ethnicity Distribution")
plt.show()
```

```python
#BMI Distribution
plt.hist(df["BMI"],bins=50,color='purple',edgecolor='black')
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
#Year wise cancer cases
df['Diagnosis_Year'].value_counts().sort_index().plot(kind='line',marker='o')
plt.title("Year-wise Cancer Cases")
plt.xlabel("Year")
plt.ylabel("Cases")
plt.show()
```
